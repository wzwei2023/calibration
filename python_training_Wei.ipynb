```python
#How to use the 'if' statement
mass = 3.54
if mass > 3.0:
    print(mass, 'is large')

mass = 2.07
if mass > 3.0:
    print(mass,' is small')
```

```python
masses = [3.54, 2.07, 1.98, 9.28]
for m in masses:
    if m > 3.0:
        print(m,'is large')
```

```python
#How to use the 'else if' statement in python
#'else if' is 'elif' in python (remember this!!)
masses = [3.54, 2.07, 1.98, 9.28]
for m in masses:
    if m > 3.0:
        print(m,'is large')
    elif m >9.0:
        print(m,' is huge')
    else:
        print(m, 'is small')
#Notes: 
# 1. python check 'if' first, if 'if' statement is true then 'elif' will not be ran by python
#  only when the 'if' statement is false, then it will check the 'elif' 
# 'else' statement means if none of the 'if' and 'elif' statement is true, then run the 'else'         
```

```python
mass = [3.54, 2.07, 9.22, 1.86, 1.71]
velocity = [10.00, 20.00, 30.00, 25.00, 20.00]

for i in range(5):
    if mass[i] > 5 and velocity[i]>20:
        print('Fast and heavy')
    elif mass[i]>2 and mass[i]<=5 and velocity[i]<=20: #can use as many 'and' as possible
        print('Normal')
    else:
        print("In the else statment")
```

```python
def average(values):
    if len(values) == 0:
        return None
    return sum(values)/len(values)
```

```python
a = average([1, 3, 4]) #the 'values' can be a list
print("average is ", a)
```
